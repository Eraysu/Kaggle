AbstractAs skin cancer is one of the most frequent cancers globally, accurate, non-invasive dermoscopy-based diagnosis becomes
essential and promising. A task of the Part 3 of the ISIC Skin Image Analysis Challenge@MICCAI 2018 is to predict seven disease
classes with skin lesion images, including melanoma (MEL), melanocytic nevus (NV), basal cell carcinoma (BCC), actinic keratosis /
Bowens disease (intraepithelial carcinoma) (AKIEC), benign keratosis (solar lentigo / seborrheic keratosis / lichen planus-like keratosis)
(BKL), dermatofibroma (DF) and vascular lesion (VASC) as defined by the International Dermatology Society [1], [2]. In this work, we
design the WonDerM pipeline, that resamples the preprocessed skin lesion images, builds neural network architecture fine-tuned with
segmentation task data (the Part 1), and uses an ensemble method to classify the seven skin diseases. Our model achieved an
accuracy of 0.899 and 0.785 in the validation set and test set, respectively.
Index TermsDermatology, Skin Lesion Classification, Computer Vision and Pattern Recognition, Deep Learning, Inductive Transfer
Learning

```python
plt.figure(figsize=(15,10))
sns.countplot(data.age)
plt.title("Age of cancer patient",size=20)
plt.show()    
```
